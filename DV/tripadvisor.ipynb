```python
import requests
from bs4 import BeautifulSoup

url = "https://pnh.cambodia-airports.aero/node/261#arrival"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Example: Scrape all headings
headings = soup.find_all("h1")
for heading in headings:
    print(heading.text.strip())
```

```python
df = pd.read_html(str(table))[0]

# Extract airline logos from the HTML
logo_links = []
for td in soup.find_all("td", class_="td-logo"):
    img_tag = td.find("img")  # Find the <img> tag
    if img_tag:
        logo_links.append(img_tag["src"])  # Extract the 'src' URL
    else:
        logo_links.append(None)
```
